```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
cancer_data = cancer.data
cancer_feature = cancer.target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
X_train, X_test, y_train, y_test = train_test_split(cancer_data, cancer_feature, test_size = 0.2)
forest = RandomForestClassifier(n_estimators = 200, random_state = 0)
forest.fit(X_train, y_train)
print('{:.3f}'.format(forest.score(X_test, y_test)))

forest = RandomForestClassifier(n_estimators = 55, random_state = 0)
forest.fit(X_train, y_train)
print('{:.3f}'.format(forest.score(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_feature_importances_cancer(model):
    n_features = cancer.data.shape[1]
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), cancer.feature_names)
    plt.xlabel("feature importance")
    plt.ylabel("feature")
    plt.ylim(-1, n_features)

plot_feature_importances_cancer(forest)
```
